```python
fname_list = glob.glob("/work/xiangchong.li/work/hsc_s23b_sim/version1/anacal_catalog_match/20250512T105446Z/simCoadd_anacal_selbias_flux_jointfpfs/*/*/*")
tracts = [int(fname.split("/")[-3]) for fname in fname_list]
patches = [int(fname.split("/")[-2]) for fname in fname_list]

tt = astTable.Table()
tt["tract"] = tracts
tt["patch"] = patches
fwhms = []
variances = []
e1s = []  
e2s = []
for dd in tt:
    tract = dd["tract"]
    patch = dd["patch"]
    fname = glob.glob("/work/xiangchong.li/work/hsc_s23b_sim/version1/dm_catalog/20250330T142355Z/sim_0_rot1_Coadd_meas/%d/%d/i/*" %(tract, patch))[0]
    data = astTable.Table.read(fname)
    data = data[data["base_Variance_value"] > 0][2]
    fwhms.append(xlens.catalog.basic.get_psf_size(data, pixel_scale=0.168))
    variances.append(data["base_Variance_value"])
    e1, e2 = xlens.catalog.basic.get_psf_ellip(data)
    e1s.append(e1)
    e2s.append(e2)
tt["fwhm"] = fwhms
tt["variance"] = variances
tt["psf_e1"] = e1s
tt["psf_e2"] = e2s
tt.write("/lustre/work/xiangchong.li/work/hsc_s23b_sim/tract_patch.csv", overwrite=True)
```

```python
tt2_list = []
for dd in tt:
    tract = dd["tract"]
    patch = dd["patch"]
    fname = glob.glob("/work/xiangchong.li/work/hsc_s23b_sim/version1/anacal_catalog_match/20250512T105446Z/simCoadd_anacal_selbias_flux_jointfpfs/%d/%d/*" %(tract, patch))[0]
    tt2_list.append(astTable.Table.read(fname)[2])
tt2 = astTable.vstack(tt2_list)
```
